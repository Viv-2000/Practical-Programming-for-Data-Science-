# Task 1: Data Preparation


```python

""""
importing all required libraries that will be used in the code
"""
import pandas as pd                        #to read files
import numpy as np
import matplotlib.pyplot as plt           #to plot graph
import warnings                           # to ignore warnings, thi swas not necessary

warnings.filterwarnings("ignore")
```

```python
#loading all provided data bases as csv files and storing them into variables

df=pd.read_csv(r'C:\Users\vivek\OneDrive\Desktop\Global database on school-age digital connectivity-Corrected\Primary.csv')
df_s = pd.read_csv(r'C:\Users\vivek\OneDrive\Desktop\Global database on school-age digital connectivity-Corrected\Secondary.csv')
df_t = pd.read_csv(r'C:\Users\vivek\OneDrive\Desktop\Global database on school-age digital connectivity-Corrected\Total School Age.csv')


df.head()         #displays first 5 rows of df, ie primary file
```

```python
#all data displayed successfully
#step 1- removing the row that has column a etc and making the 0th row are the header row
# For primamry
df.columns = df.iloc[0]                   
df.drop(index=df.index[0], axis=0,inplace=True)    #drop function deletes the respective row

# For secondary
df_s.columns = df_s.iloc[0]
df_s.drop(index=df_s.index[0], axis=0,inplace=True)

# For total school age
df_t.columns = df_t.iloc[0]
df_t.drop(index=df_t.index[0], axis=0,inplace=True)
```

```python
#displaying shape of the data 
print("primary")
print(df.shape,'\n')      #shape function gives the shape as a tuple of number of rows and columns

print("secondry")
print(df_s.shape,'\n')

print("Total Shool Age")
print(df_t.shape)
```

```python
#printing the unique values so that we can remove the invalid values for primary 
print("Primary")
print(df['Time period'].unique())
print(df['Sub-region'].unique())
print(df['Region'].unique(),'\n')

#printing the unique values so that we can remove the invalid values for secondary
print("Secondary")
print(df_s['Time period'].unique())
print(df_s['Sub-region'].unique())
print(df_s['Region'].unique(),'\n')

```

```python
# dropping invalid data from primamry 
df.drop(df[df['Time period'].isin(['2562', '2027', '2012-99', '2076'])].index, inplace = True)

# dropping invalid data from Secondary
df_s.drop(df_s[df_s['Time period'].isin(['2562', '3019','2012-99','2076'])].index, inplace = True)
```

```python
df.shape
```

```python
# Joining "ISO3" and "Countries and areas" into in column "Country"  
# Primamry
df['Country'] = df[['ISO3', 'Countries and areas']].apply(lambda x: '({}) {}'.format(x[0], x[1]), axis=1)

# DOing same for Secondary  
df_s['Country'] = df_s[['ISO3', 'Countries and areas']].apply(lambda x: '({}) {}'.format(x[0], x[1]), axis=1)
```

```python
df.shape
```

```python
# Remove four invalid columns
# Primamry
df.drop(df.columns[[0, 1, 2, 10]], axis=1, inplace=True)

# Secondary
df_s.drop(df_s.columns[[0, 1, 2, 10]], axis=1, inplace=True)
```

```python
# checking by shape whether data is updated 
# Primamry
print(df.shape)

# Secondary
df_s.shape
```

```python
#checking in which columns null values are present.

print(" Null Values:" )

df.isna().sum()
```

```python
#renaming column names 
# for primary
df.rename(columns = {'Sub-region':'Region','Total':'Total_%','Rural (Residence)':'Rural_%', 'Urban (Residence)' : 'Urban_%', 'Poorest (Wealth quintile)':'Poorest_%', 'Richest (Wealth quintile)':'Richest_%' , 'Time period':'Survey year'}, inplace = True)

# for Secondary
df_s.rename(columns = {'Sub-region':'Region','Total':'Total_%','Rural (Residence)':'Rural_%', 'Urban (Residence)' : 'Urban_%', 'Poorest (Wealth quintile)':'Poorest_%', 'Richest (Wealth quintile)':'Richest_%' , 'Time period':'Survey year'}, inplace = True)
```

```python
#locating all null values in the following columns (Rural, Urban, Poorest, Richest)

null_values = df[['Rural_%', 'Richest_%', 'Urban_%', 'Poorest_%']].isna().all(axis=1)
df.loc[null_values]
```

```python
#locating all null values in all 4 columns (Rural, Urban, Poorest, Richest)
# Secondary
null_values = df_s[['Rural_%', 'Richest_%', 'Urban_%', 'Poorest_%']].isna().all(axis=1)
df_s.loc[null_values]
```

```python
# Drop rows with missing values

df.dropna(subset=['Rural_%', 'Richest_%', 'Urban_%', 'Poorest_%'], how = "all" ,inplace=True)
df_s.dropna(subset=['Rural_%', 'Richest_%', 'Urban_%', 'Poorest_%'], how = "all" ,inplace=True)
print(df.shape,df_s.shape)
```

```python
# data types of all columns  
df.info()
```

```python
# changing numerical columns into float data type so that our exploration part becomes easier 

df['Total_%'] = df['Total_%'].str.replace('%', '').astype(float)
df['Poorest_%'] = df['Poorest_%'].str.replace('%', '').astype(float)
df['Urban_%'] = df['Urban_%'].str.replace('%', '').astype(float)
df['Rural_%'] = df['Rural_%'].str.replace('%', '').astype(float)
df['Richest_%'] = df['Richest_%'].str.replace('%', '').astype(float)


df_s['Total_%'] = df_s['Total_%'].str.replace('%', '').astype(float)
df_s['Poorest_%'] = df_s['Poorest_%'].str.replace('%', '').astype(float)
df_s['Urban_%'] = df_s['Urban_%'].str.replace('%', '').astype(float)
df_s['Rural_%'] = df_s['Rural_%'].str.replace('%', '').astype(float)
df_s['Richest_%'] = df_s['Richest_%'].str.replace('%', '').astype(float)
```

```python
# displaying changed data into float 
df.head()
```

```python
# extract year from Survey year column
df['Year'] = pd.to_datetime(df['Survey year'].str[:4], format='%Y').dt.year

# drop Survey year 
df = df.drop(columns=['Survey year'])


# extract year from Survey year column 
df_s['Year'] = pd.to_datetime(df_s['Survey year'].str[:4], format='%Y').dt.year

# drop Survey year column
df_s = df_s.drop(columns=['Survey year'])
```

```python
#using reindex 
df = df.reindex(columns=['Country', 'Region', 'Income Group','Total_%','Rural_%','Urban_%','Poorest_%','Richest_%','Year'])
df_s = df_s.reindex(columns=['Country', 'Region', 'Income Group','Total_%','Rural_%','Urban_%','Poorest_%','Richest_%','Year'])
```

```python
#Nominal data 
income_groups = df.groupby("Income Group")["Country"].count()

# Plot the bar chart of the distribution of countries by income group
# The graph calculates the frequency of countries in each income group using
plt.figure(figsize=(15, 8))
income_groups.plot(kind="bar")
plt.title("Distribution of Countries by Income Group")
plt.xlabel("Income Group")
plt.ylabel("Number of Countries")
plt.show()
```

```python
# ordinal value

# count number of occurrences of each value in the time_period column
time_period_counts = df['Year'].value_counts()

# convert the index (time periods) to a list
time_periods = time_period_counts.index.tolist()

# convert the counts to a list
count_time_period = time_period_counts.tolist()

# plot bar chart
plt.figure(figsize=(15, 8))
plt.bar(time_periods, count_time_period)
plt.xticks(time_periods)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Maximum and minimum data gathered')
plt.show()
```

```python
#Numerical data 
plt.boxplot(df['Urban_%'])
plt.xlabel('poorest (in percentage)')
plt.title('Percentage of Urban liveing Children in Primary Schools')
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df,alpha=0.8,figsize=(16,16),diagonal='density')
plt.show()
```
